```python
while True:
    random_index = random.randint(0,len(strings)-1)
    string = strings[random_index]
    word_count = len(string.split())
    print(string)
    t0 = time.time()
    input_txt = str(input('Enter the sentence: '))
    t1 = time.time()
    
    if input_txt == '0':
        print('Thank you for choosing to test typing speed with us!')
        break
        
    correct_words = len(set(input_txt.split())&set(string.split()))
    accuracy = correct_words/word_count
    time_taken = t1-t0
    wps = word_count/time_taken
    wpm = wps*60
    print('WPS: ', wps, '\nWPM: ', wpm, "\nAccuracy: ", accuracy*100, "\nTime Taken: ", time_taken, "\n\n")
```
